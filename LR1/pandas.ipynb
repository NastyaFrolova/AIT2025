```python
# Для следующего пункта я решила сделать новый столбец, отображающий связь между уровнем сахара в крови и выработкой инсулина. Таким образом, более высокое соотношение может указывать на резистентность к инсулину
# Есть строки с инсулином = 0, что говорит о необходимости замены на малое число, таким образом в G/I ratio выполнена замена inf на 0, так как итоговый результат будет близок к 0

data = df.copy()
data['G/I ratio'] = (data["Glucose"] / data["Insulin"]).round(2)
data['G/I ratio'] = data['G/I ratio'].replace(np.inf, 0)
print(data)
```

через DataFrame.apply;

```python
# Создание нового поля на основе BMI через DataFrame.apply

def bmi_category(row):
    bmi = row['BMI']
    if bmi < 18.5:
        return "Underweight"
    elif 18.5 <= bmi < 24.9:
        return "Normal weight"
    elif 25 <= bmi < 29.9:
        return "Overweight"
    else:
        return "Obese"

data['BMICategory'] = data.apply(bmi_category, axis=1)
print(data)
```

через Series.apply.

```python
# Создание нового поля на основе BloodPressure через Series.apply

def bp_classification(bp):
    if bp < 120:
        return "Normal"
    elif 120 <= bp < 139:
        return "Elevated"
    elif 140 <= bp < 159:
        return "Stage 1 Hypertension"
    else:
        return "Stage 2 Hypertension"

data['BloodPressureClassification'] = data['BloodPressure'].apply(bp_classification)
print(data)
```

Решейпинг данных 1D→2D с использованием .pivot (можно подать на вход результаты агрегатов, полученных ранее с использованием .groupby (сгруппировать по двум полям)).

```python
# Выполним группировку по BMI и возрасту, заполняя средним значением Outcome

df_for_pivot = df.groupby(['Pregnancies', 'Age'], as_index=False)['BMI'].mean()
print(df_for_pivot)
```

```python
df_pivot = df_for_pivot.pivot(columns='Age', index='Pregnancies', values='BMI').fillna('-//-')
df_pivot
```
